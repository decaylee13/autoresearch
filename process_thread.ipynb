```python
# Using Time Module in Python
import time

time.sleep(2)
curr_time = time.time()
print(f"Current time: {curr_time}")

```

```python
# Using Threading Module in Python
import threading 

def sample_method(name): 
    print(f"Thread {name} is starting.")
    time.sleep(2)
    print(f"Thread {name} is ending.")

t = threading.Thread(target=sample_method, args=("dk"))
t.start()
t.join()
```
